# My Solution

We are told this is an unsupervised clustering problem - there are many methods to use. This solution uses KMeans clustering from the scikit learn library. I chose this because it is a well understood algorithm, robust, and fast. I understand the shortcomings of KMeans to be that it is somewhat biased towards ellipsoidal clusters. If we have clustering that is non-ellipsoidal, then we will not get a very good result from our cluster analysis.

Another clustering solution I wanted to try, but did not have time to implement was DBScan. I understand that this is more robust, with regards to handling oddly shaped clusters, and it would provide an important cross-check to KMeans.

A notorious problem with unsupervised learning is validation - which I have attempted to qualitatively do by studying the inertia of clusters as a funciton of number of clusters. There is a slight elbow in this distribuiton, which suggests that the optimal number of clusters for kMeans is somewhere between 2 and 4.

Assuming that the each driver for the car does not drastically change their driving behavior, we make the ansatz that the number of clusters correlates to the number of drivers - but this might actually not be true. For example - some trips might occur during rush hour, and other trips might occur on a clear freeway - so without labled features, it is hard to study whether or not clustering will be more correlated ot driving conditions, or to number of drivers.

## Feature Correlation - Dimensionality Reduction
Although this data set has 81 features, we can still use KMeans without worry of overburdening our computer. However, as a consistency check, for perhaps trying the same clustering algorithms in reduced dimensional space, we can do a sanity check to determine how correlated varaibles are.

So, the first step is to check for corellated parameter pairs, in order to determine if feature reduction will be a good appproach without losing too much analyzing power. We can use this informaiton later if run-time is too large for k-means.

Grouping Correlated parameters might lead to some insights.

On the plot below, black is highly correlated, and white is uncorrelated The black diagonal line represents 100% correlation (pearson), which is simply a variable's correlation with itself.

```python
from math import fabs

p_corr = df.corr(method='pearson')
correlations = p_corr.values
plt.pcolor(p_corr)
plt.figure(figsize=(20,20))
plt.show()
```

## Correlations, continued
Another possible direction is to look at features which are "highly correlated" and look at features which are not correlated. The thresholds can be set arbitrarily.

```python
correlations_vs_index = {}
for i in range(1,len(correlations)):
    for j in range(i+1,len(correlations)):
        if i == j: continue
        index_string = str(i)+'_'+str(j)
        # since i,j correpsond to the same corellation pair, we don't need to worry about
        # overwriting. Additionally, for sorting purposes, we take the absolute value of the
        # correlation.
        correlations_vs_index[index_string] = fabs(correlations[i][j])
sorted_keys = sorted(correlations_vs_index,key=correlations_vs_index.get)

# Count the number of variables above/below the thresholds
low_correlation = 0
high_correlation = 0

# Correlation thresholds
high_correlation_threshold = 0.9
low_correlation_threshold = 0.1

correlated_keys = []
uncorrelated_keys = []
for key in sorted_keys:
    if correlations_vs_index[key] > high_correlation_threshold:
        high_correlation += 1
        correlated_keys.append(key)
    if correlations_vs_index[key] < low_correlation_threshold:
        low_correlation += 1
        uncorrelated_keys.append(key)
print high_correlation,'variable pairs were highly correlated ( >',high_correlation_threshold,'Pearson Coefficent)'
print low_correlation,'variable pairs were not very correlated ( <',low_correlation_threshold,'Pearson Coefficient)'
```

# Clustering with KMeans

Kmeans is robust, but makes some assumptions about cluster shape which may not be valid. Generally, kmeans is biased towards ellipsoidal clusters. We can get an estimate for the number of drivers represented in this data set (subject to the caveats mentioned regarding driving conditions, and the limitations of k-means), by running k-means clustering many times, and looking at the variance accounted for by the clustering.

```python
from sklearn.cluster import KMeans
from scipy.cluster.vq import kmeans
# Convert DataFrame of data to Matrix, X
X = df.as_matrix()

# Try a few different clustering sizes
kmeans = {}
clusters = {}
results = {}
inertia = {}
for i in range(1,10):
    kmeans[i] = KMeans(n_clusters=i)
    kmeans[i].fit(X)
    clusters[i] = kmeans[i].labels_
    inertia[i] = kmeans[i].inertia_
    results[i] = pd.DataFrame([df.index,clusters[i]]).T

```

# Estimating the number of drivers
We can plot the inertia (the distance between centroids and data points for each cluster) as a function of the number of clusters.

The elbow of the distribution suggests the optimum number of clusters, which we can estimate as the number of drivers for the car.

In this case, we see a slight elbow at approximately 2-4 clusters, which corresponds to 2-4 unique drivers for this car.

While certainly qualitatively, this is a reasonable number to assume for a family, it is by no means exhaustively correct.

The number of apparent clusters might be affected by the geometry of the clusters themselves, so other clustering metrics should be use to further study.

```python
import matplotlib.pyplot as plt

inertia_list = []
clusters = []
for i in range(1,10):
    inertia_list.append(inertia[i])
    clusters.append(i)
x_vals = np.array(clusters)
y_vals = np.array(inertia_list)
plt.scatter(x_vals,y_vals)
plt.show()
```

# Attempting to Cluster Data With DBScan
K-Means clustering requires that we know the exact number of clusters, as well as making some assumptions about the general shape of clusters. We don't really know that clusters will be ellipsoidal.

To use DB Scan, we have to define an appropriate distance metric to describe the distance between observations, which are represented in our dataframe rows. Once we have defined this distance metric, then we can proceed with clustering the data.

This method has not yet been implimented, but it would serve as an important cross-check to KMeans.

We examine the distribution of distances below:

```python
from scipy.spatial.distance import pdist

# All possible pairs of distances
distance_matrix = pdist(df)
```
