# Primer on NLP

```python
%matplotlib notebook
import re
import math
import string
from collections import Counter
```

```python
TEXT = open('big.txt').read()
```

```python
print(len(TEXT))
```

```python
def tokens(text):
    """List all the word tokens (consecutive letters) in a text. Normalize to lowercase"""
    return re.findall('[a-z]+',text.lower())
```

```python
WORDS = tokens(TEXT)
```

```python
print(WORDS[:10])
```

```python
def sample(bag,n=10):
    """Samples 10 words randomly from bag"""
    return ' '.join(random.choice(bag) for _ in range(n))
```

```python
sample(WORDS)
```

```python
COUNTS = Counter(WORDS)
print(COUNTS.most_common(10))
```

```python
for w in tokens('the rare and neverbeforeseen words'):
    print(COUNTS[w],w)
```

```python
M = COUNTS['the']
```

```python
print(M)
```

```python
M = COUNTS['the']
yscale('log');xscale('log'); title('Frequency of the n-th most frequent word and 1/n line.')
plot([c for (w,c) in COUNTS.most_common()])
plot([M/i for i in range(1,len(COUNTS)+1)])
```
